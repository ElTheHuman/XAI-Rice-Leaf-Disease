# **Exploratory Data Analysis**

# Set directory path

```python
import os
```

```python
train_aug_dir = '../data/Rice_Leaf_AUG/Rice_Leaf_AUG'
class_list = os.listdir(train_aug_dir)
```

# Number of classes

```python
print(f'Number of classes: {len(class_list)}')
```

# Number of records in every class

```python
total_records = 0
num_records = []

for classes in class_list:
    subdir = os.path.join(train_aug_dir, classes)
    curr_num_records = len(os.listdir(subdir))
    print(f'Number of records in class {classes}: {curr_num_records}')
    num_records.append(curr_num_records)
    total_records += curr_num_records
```

# Total records

```python
print(f'Total records for all classes: {total_records}')
```

# Visualization

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,6))
plt.pie(num_records, labels=class_list, autopct='%1.2f%%')
plt.show()
```

# Conclusion

The classes records are almost evenly distributed based on the pie chart
