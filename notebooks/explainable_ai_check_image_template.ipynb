# Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Define Deprocess Function

```python
def deprocess_resnet(x):
    """Deprocess image that used keras.applications.resnet.preprocess_input"""
    x = x.copy()
    mean = [103.939, 116.779, 123.68]  # BGR, From ImageNet mean
    # Return values using mean
    x[..., 0] += mean[0]
    x[..., 1] += mean[1]
    x[..., 2] += mean[2]
    # Reverse BGR tp RGB
    x = x[..., ::-1]
    # Make sure that the values are in [0-255]
    return np.clip(x, 0, 255).astype("uint8")
```

# Load All Image Array

Default images are processed using resnet preprocessing

```python
image_array = np.load('../src//image_array/image_array.npy')
occluded_image_array = np.load('../src/image_array/occluded_image_array.npy')
```

# Check Normal Processed Image

```python
for img in image_array:
    plt.imshow(img.astype('uint8'))
    plt.axis('off')
    plt.show()
```

# Check Normal Original Image

```python
for img in image_array:
    plt.imshow(deprocess_resnet(img))
    plt.axis("off")
    plt.show()
```

# Check Occluded Processed Image

```python
for img in occluded_image_array:
    plt.imshow(img.astype('uint8'))
    plt.axis('off')
    plt.show()
```

# Check Occluded Original Image

```python
for img in occluded_image_array:
    plt.imshow(deprocess_resnet(img))
    plt.axis("off")
    plt.show()
```
